```python
import yfinance as yf
```

```python
##google hisse verilerini 2020-2025 arası çekti df ile de gösterdi bize
google=yf.Ticker("GOOGL")
df=google.history(start="2020-01-01",
                  end="2025-07-11")
df
```

```python
df['Close'].plot(figsize=(10,5))  ##çekilen verileri görselleştiriyor
```

```python
nvda=yf.Ticker("NVDA")
df=nvda.history(start="2020-01-01",
                end="2025-07-11") ## nvida hisse verilerini 2020-2025 arası çekiyor
```

```python
df['Close'].plot()
```

```python
##max değerini buluyoruz
df['Close'].max()
```

```python
##2 veriyi karsılaştırmak için download kullandık
df2=yf.download(["NVDA","INTC"],
                start="2015-01-01",
                end="2025-07-11")
```
