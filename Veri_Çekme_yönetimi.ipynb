```python
##webten bir tablodan veri çekme
import pandas as pd
df=pd.read_html(r"https://en.wikipedia.org/wiki/List_of_countries_by_number_of_scientific_and_technical_journal_articles")[1] ##r atmamızın nedeni linki tam olarak alsın kaçacagı yer olmasın diye
df.head()
```

```python
##veri seti hakkında bilgi al totali kullanamadın
df.info()
```

```python
##kolon isimlerini değiştiriyoruz
df.columns=['Rank','Country', 'Total', 'ApM']
df.head()
```

```python
##üstteki infoya göre kolon isimlerin değiştirdik
df.info()
```

```python
##ülkerleri toplam yayın sayısına göre sıralama
df.sort_values(by="Total",ascending=False, inplace=True)
df.head(10)
```

```python
df.head(10).plot(x="Country",y="Total",kind="bar")
```

```python
df2=pd.read_html(r"https://tr.wikipedia.org/wiki/N%C3%BCfuslar%C4%B1na_g%C3%B6re_%C3%BClkeler_ve_ba%C4%9F%C4%B1ml%C4%B1_topraklar_listesi")[1]
print(df2.head())
# Nüfus sütunun sayıya çevrilmesi
df2['Nüfus'] = df2['Nüfus'].str.replace('.', '', regex=False) ##ondalık nokta ayraçlarını kaldırıyorsun burda
df2['Nüfus'] = pd.to_numeric(df2['Nüfus']) ##obj de numeric yapıyorsun
df2.sort_values(by='Nüfus',ascending=False,inplace=True) ## sıralamasını yapıyorsun
print(df2.info())
df2=df2.iloc[1:10].plot(x='Ülke',y='Nüfus',kind='bar')
```
